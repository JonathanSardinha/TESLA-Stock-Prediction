```python
tesla=tesla.fillna(tesla.mean())
tesla
```

```python
# Defining Predictor Variable

tesla['Open-Close'] = tesla.Open - tesla.Close
tesla['High-Low'] = tesla.High - tesla.Low

X = tesla[['Open-Close', 'High-Low']]
X.head()
```

```python
# Defining Target Variable
# +1: Buy signal, -1: Sell signal
Y = np.where(tesla['Close'].shift(-1)>tesla['Close'],1,-1)
```

```python
# Splitting the dataset in 70-30 ratio
split_percentage = 0.7
split = int(split_percentage*len(tesla))

X_train = X[:split]
Y_train = Y[:split]

X_test = X[split:]
Y_test = Y[split:]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Initiating KNN classifier
knn = KNeighborsClassifier(n_neighbors=15)

# fit the model
knn.fit(X_train,Y_train)

# Accuracy
accuracy_train = accuracy_score(Y_train, knn.predict(X_train))
accuracy_test = accuracy_score(Y_test, knn.predict(X_test))

print('Train_data accuracy: %2f' %accuracy_train)
print('Test_data accuracy: %2f' %accuracy_test)
```

```python
# Predicted Signal
tesla['Predicted Signal'] = knn.predict(X)

# SPY Cumulative returns
tesla['SPY_returns'] = np.log(tesla['Close']/tesla['Close'].shift(1))
Cumulative_SPY_returns = tesla[split:]['SPY_returns'].cumsum()*100

#Cumulative Strategy returns
tesla['Strategy_returns'] = tesla['SPY_returns']* tesla['Predicted Signal'].shift(1)
Cumulative_Strategy_returns = tesla[split:]['Strategy_returns'].cumsum()*100

#plotting results
plt.figure(figsize=(10,5))
plt.plot(Cumulative_SPY_returns,color='r',label='SPY Returns')
plt.plot(Cumulative_Strategy_returns,color='g',label='Strategy Returns')
plt.legend()
plt.show()
```
